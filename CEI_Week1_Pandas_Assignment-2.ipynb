```python
import pandas as pd

data = {
    "product": ["Soap", "Shampoo", "Face Wash", "Soap", "Lotion"],
    "price": [50, 150, 120, 50, None],
    "quantity": [2, 1, 3, 2, 4]
}

df = pd.DataFrame(data)

df.to_csv("sales.csv", index=False)

print("sales.csv created")
```

```python
import pandas as pd

df = pd.read_csv("sales.csv")
df
```

```python
print("Head:")
print(df.head())

print("\nTail:")
print(df.tail())

print("\nShape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
df["price"] = df["price"].fillna(df["price"].mean())

df
```

```python
df[["product", "price"]]
```

```python
filtered_df = df[df["price"] > 100]

filtered_df
```

```python
df = df.drop_duplicates()

df
```

```python
df["total_amount"] = df["price"] * df["quantity"]

df
```

```python
df.to_csv("cleaned_sales.csv", index=False)

print("cleaned_sales.csv saved")
```
